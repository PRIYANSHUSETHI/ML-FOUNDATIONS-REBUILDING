UNDERSTANDING YOUR DATA. ASKING BASIC QUESTIONS TO KNOW MORE ABOUT WHAT THE DATA LOOKS LIKE AND WHAT IT REPRESENTS.

Step - 1: Loading the Data

```python
import pandas as pd
df = pd.read_csv('titanic.csv')
```

Question-1: How big is your data?

This tells us the number of rows and coloumns.

```python
df.shape
```

Question-2: How does data look like?

The head shows the top 5 rows and all the coloumns of the dataset. But this might not be the best way to judge your data as there may be a bias in the starting or the end of the dataset.

```python
df.head()
```

To make sure that you do not see the biased dataset, use sapmle() function of the pandas library, this randomly selects 5 data points to show it to you.

```python
df.sample(5)
```

Question-3: What is the data type of each coloumn of my dataset?

```python
df.info()
```

Question-4: Are there any missing values?

```python
df.isnull().sum()
```

Question-5: How does data look mathematically?

The describe function gives me a high level mathematical summary of data in each coloumn. The only problem to this is that the insights that you will get will be only from the numerical coloumns.

```python
df.describe()
```

Question-6: Are there duplicate values?

```python
df.duplicated().sum()
```

Question-7: How is the correlation between coloumns?

```python
df.corr(numeric_only= True)
```

If we want correlation between all the coloumns and one couloumn in particular we write like the following:

```python
df.corr(numeric_only = True)['Survived']
```
